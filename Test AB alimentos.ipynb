A menudo necesitarás analizar embudos de eventos, que tienen similitudes con los embudos de marketing y productos que ya conoces. En el análisis basado en eventos, un embudo es una determinada serie de acciones que los usuarios y las usuarias deben completar en un sitio web, en un juego o en una aplicación para lograr un determinado resultado.

Los embudos para las tiendas en línea son bastante simples: no puedes pagar hasta que agregues un producto al carrito y, en realidad, no puedes comprar nada sin completar el proceso de pago.

Las aplicaciones móviles también tienen embudos. Digamos que hay una aplicación llamada Kittygram. En ella los usuarios y las usuarias pueden publicar fotos de gatos y ver fotos de los gatos de otros usuarios o usuarias. Construyamos un embudo:
* El usuario descarga la aplicación. 
* El usuario se registra. 
* El usuario completa su perfil. 
* El usuario realiza su primera acción significativa, ya sea publicando su primera foto o siguiendo a otro usuario.

Kittygram rastrea varias métricas clave:
* La proporción de usuarios registrados entre los que descargan la aplicación.
* La proporción de usuarios registrados que completan sus perfiles entre el total de usuarios registrados.
* La proporción de usuarios que realizan su primera acción significativa (publicar una foto o seguir a otro usuario) entre los usuarios que completan sus perfiles.

La última etapa de este embudo tiene dos ramas. Sería interesante saber cuántos usuarios siguen a otros primero y cuántos publican fotos de gatos primero. Esto podría potencialmente revelar la existencia de dos tipos de usuarios: creadores de contenido y consumidores de contenido.

Otra métrica importante es el tiempo entre eventos. Por ejemplo, sabemos que la mediana de tiempo hasta la primera publicación es de siete días. Si queremos que esa cifra sea menor, puede ser una buena idea enviar notificaciones push recordando a los usuarios que pueden compartir fotos de gatos en la aplicación. Pero ten cuidado de no enviar demasiadas notificaciones o los usuarios y las usuarias desactivarán las notificaciones y tu plan fracasará.

Es importante no solo calcular las métricas para el progreso a través del embudo (tasas de conversión de una etapa a otra), sino también realizar un seguimiento de los usuarios y las usuarias que no llegaron a una determinada etapa. Puede haber toda una clase de usuarios que no sigan otras cuentas ni publiquen nada. ¿No tienes curiosidad por saber qué están tramando? ¿Será que admiran en secreto a los gatos de otras personas?

#### Utiliza el embudo de eventos para encontrar la proporción de usuarios que pasan de una etapa a la siguiente. Por ejemplo, para la secuencia de eventos A → B → C, calcula la proporción de usuarios en la etapa B a la cantidad de usuarios en la etapa A y la proporción de usuarios en la etapa C a la cantidad en la etapa B.

```python
# Embudo siguiendo los eventos en órden consecutivo

# Creamos tabla pivote 
pivot_logs = df_logs_upd.pivot_table(
    index = 'user_id',                 # El índice es el 'user_id'
    columns = 'event',                 # Los 'events' son las columnas
    values = 'date_time',              # Agrega la fecha y hora en la que se dio el evento
    aggfunc = 'min'                    # Agrega la fecha mínima en cada valor (en caso de que el usuario tenga actividades repetidas durante la prueba)
)

# Filtros para crear el embudo
main_screen = ~pivot_logs['MainScreenAppear'].isna()                                                # True o False si usuario accedió a 'main_screen'
offer_screen = main_screen & (pivot_logs['OffersScreenAppear'] > pivot_logs['MainScreenAppear'])    # True o False si usuario accedió después a 'offer_screen
cart_screen = offer_screen & (pivot_logs['CartScreenAppear'] > pivot_logs['OffersScreenAppear'])    # True o False si usuario accedió después a 'cart_screen'
pay_screen = cart_screen & (pivot_logs['PaymentScreenSuccessful'] > pivot_logs['CartScreenAppear']) # True o False si usuario accedió después a 'pay_screen'

n_ms = pivot_logs[main_screen].shape[0]     # Número de True para main_screen 
n_os = pivot_logs[offer_screen].shape[0]    # Número de True para main -> offer
n_cs = pivot_logs[cart_screen].shape[0]     # Número de True para main -> offer -> cart
n_ps = pivot_logs[pay_screen].shape[0]      # Número de True para main -> offer -> cart -> pay

# Creamos un dataframe con los datos recoplidados
df_funnel = pd.DataFrame({
    'event': ['main_screen', 'offer_screen', 'cart_screen', 'pay_screen'],
    'n_users': [n_ms, n_os, n_cs, n_ps]
})

df_funnel
```

```python
# Gráfico para embudo Pantalla Principal -> Oferta -> Carrito -> Pago

sns.lineplot(data = df_funnel, x = 'event', y = 'n_users', marker = 'o')

plt.title('Embudo Pantalla Principal > Oferta > Carrito > Pago')
plt.xlabel('Evento')
plt.ylabel('# Número de usuarios')

plt.show()
```

Tomamos la primer actividad registrada por un usuario único y lo sometemos al embudo. El gráfico muestra el flujo que tuvieron el usuario a través del embudo completo  **Pantalla Principal -> Oferta -> Carrito -> Pago**. Se puede apreciar que el número de usuarios va decreciendo a medida que se pasa al siguiente evento. Inicialmente tenemos 7419 usuarios que ingresaron a la pantalla principal, de los cuales solamente 454 siguió el embudo planteado hasta llegar a la compra exitosa.

#### Tenemos dos grupos de control en el test A/A, donde comprobamos nuestros mecanismos y cálculos. Observa si hay una diferencia estadísticamente significativa entre las muestras 246 y 247.

```python
# Comparar número de compras por usuario en grupo 246 y 247

# Separamos los datos en el primer grupo A (246) y el segundo A (247) 
a_df_logs = df_logs_upd[df_logs_upd['group'] == 246]
aa_df_logs = df_logs_upd[df_logs_upd['group'] == 247]

# Escogemos aquellos eventos donde se hayan registrado compras exitosas
a_purchases = a_df_logs[a_df_logs['event'] == 'PaymentScreenSuccessful']
aa_purchases = aa_df_logs[aa_df_logs['event'] == 'PaymentScreenSuccessful']

# Agrupamos por grupo y user_id para saber cuántas compras exitosas realizó un usuario del grupo durante la prueba
user_purchases_a = a_purchases.groupby(['group', 'user_id']).size().reset_index(name='purchase_n')
user_purchases_aa = aa_purchases.groupby(['group', 'user_id']).size().reset_index(name='purchase_n')

# Creamos el mannwhitney para encontrar la distribución de compras por usuario por grupo  
user_prc_result = st.mannwhitneyu(user_purchases_a['purchase_n'], user_purchases_aa['purchase_n'])

alpha = 0.05            # Alfa de 0.05 

print(f'valor p: {user_prc_result.pvalue:.5f}')

if (user_prc_result.pvalue < alpha):
    print("Hipótesis nula rechazada: La distribución de los grupos de control (A/A) es diferente")
else: 
    print("No se rechaza la hipótesis nula: No se encontró diferencia significativa entre las distribuciones de los grupos de control (A/A)")
```

Para compobar que no hay diferencia significativa en el test A/A se optó por comparar el número de compras por usuario en el grupo 246 (primer A) y el 247 (segundo A). El test Mann-Whitney compara distribuciones de valores individuales entre ambos grupos y se decidió compara las compras por usuario de ambos grupos. 

El resultado del valor p resultó ser mayor a la significancia definida (alfa = 0.05) por lo que no hay elemenots para rechazar la hipótesis nula que supone la igualdad entre ambos grupos. Podemos proseguir con la prueba una vez demostrado que ambos grupos de control mantienen rangos similares.

```python
# Función para proporción 

def event_proportion (df, event):                                       # La función requiere de un 'dataframe' y un 'evento'
    df_event = df[df['event'] == event]                                 # Filtramos el 'dataframe' por el nombre del 'evento'

    users_per_event = df_event.groupby('group')['user_id'].nunique()    # Agrupamos por grupo y sacamos el número de usuarios que hicieron ese evento
    tot_users_grp = df.groupby('group')['user_id'].nunique()            # Obtenemos el número total de usuarios del grupo

    prop_event = users_per_event/tot_users_grp                          # Sacamos la proporción

    return prop_event                                                   # Regresa la proporción 
```

```python
# Evaluamos la función de proporción por evento

results_prop = []                                                       # Creamos lista que guarde los resultados

for event in df_logs_upd['event'].unique():                             # Para cada 'event' único en df_logs_upd...
    prop_event = event_proportion(df_logs_upd, event)                   # Llamamos a la función 
    results_prop.append({                                               # Excluyendo al grupo B (248), agregamos a la lista que guarda los resultados...
        'event': event,                                                 # El nombre del evento
        'prop_246': prop_event.get(246),                                # Proporción del grupo 246
        'prop_247': prop_event.get(247)                                 # Proporción del grupo 247   
    })

df_results_prop = pd.DataFrame(results_prop)                            # Pasamos la lista a un dataframe y los imprimimos

print(df_results_prop)
```

A primera instancia, la diferencia entre las proporciones del primer grupo A (246) vs el segundo grupo A (247) no parecen ser muy diferentes, siendo la mayor diferencia durante el evento 'PaymentScreenSuccessful' con una diferencia de 2.3%. 

Realizamos la prueba Mannwhitney para comparar las distribuciones.

```python
# ¿Diferencia entre grupos estadísticamente significativa para cada evento? Función


def event_distr (df, event, test_1, test_2):
    df_event = df[df['event'] == event]                                 # Filtramos el 'dataframe' por el nombre del 'evento'

    # Obtenemos el número de veces que un usuario único  generó para cada evento 
    # Filtramos el dataframe por grupo (246, 247, 248), agrupamos por 'user_id' y sacamos el tamaño (número de veces que aparece) y le asignamos el conteo a la columna 'n_event'
    test1_users_event = df_event[df_event['group'] == test_1].groupby('user_id').size().reset_index(name='n_event')    
    test2_users_event = df_event[df_event['group'] == test_2].groupby('user_id').size().reset_index(name='n_event')
    
    # Sacamos el valor p de la prueba Mannwhitney comprando el número de veces que el usuario realizó determinado evento
    stat, p_value = st.mannwhitneyu(test1_users_event['n_event'], test2_users_event['n_event'])

    return p_value                                                      # Devolvemos el valor p
```

```python
# Evaluamos la función de Mannwhitney 

result_distr_event = []                                                 # Creamos lista para almacenar los resultados

grp_a = 246                                                             # Asignamos el primer grupo A
grp_aa = 247                                                            # Asignamos el segundo grupo A

alpha = 0.05                                                            # Significancia de 0.05

for event in df_logs_upd['event'].unique():                             # Para cada 'event' único en df_logs_upd...
    p_value = event_distr(df_logs_upd, event, grp_a, grp_aa)            # Llamamos a la función asignando ambos grupos A
    result_distr_event.append({                                         # Agregamos los resultados a la lista
        'event': event,                                                 # El nombre del evento
        'p_value': p_value                                              # El valor p de la prueba
    })
    if p_value < alpha:                                                 # Evaluamos la hipótesis para la distribución de cada evento
        print(f'valor p: {p_value:.5f}')
        print(f'Para {event}: Hipótesis nula rechazada: La distribución de los grupos de control (A/A) es diferente')
    else:
        print(f'valor p: {p_value:.5f}')
        print(f'Para {event}: No se rechaza la hipótesis nula: No se encontró diferencia significativa entre las distribuciones de los grupos de control (A/A')
```
